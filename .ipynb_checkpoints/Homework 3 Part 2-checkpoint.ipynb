1. (8 points) **Solve for the parameters ($\mathbf{w}$ and $w_0$) of the discriminant function you solved for in HW3-Part1-Q1. Plot the corresponding discriminant function.**

````python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
X = np.load('data.npy')
t = np.load('labels.npy')

# Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=t)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Dataset Scatter Plot')
plt.show()
````

````python
from sklearn.linear_model import LinearRegression

# Fit a linear model using least squares
model = LinearRegression(fit_intercept=True)
model.fit(X, t)

# Extract the coefficients
w = model.coef_
w_0 = model.intercept_

print("Weights:", w)
print("Bias:", w_0)
````

````python
def plot_decision_boundary(w, w_0):
    # Create a grid of points
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    # Calculate the decision boundary
    Z = w[0] * xx + w[1] * yy + w_0
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.colorbar()

    # Plot the data points
    plt.scatter(X[:, 0], X[:, 1], c=t)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Decision Boundary')
    plt.show()

plot_decision_boundary(w, w_0)
````

2. (2 points) **Compare your solution with the standard LDA discriminant function.**

````python
g(x) = w^T x + w_0

w = Σ_1^-1 (μ_1 - μ_2)
w_0 = -0.5 (μ_1^T Σ_1^-1 μ_1 - μ_2^T Σ_2^-1 μ_2) - ln(p_1 / p_2)

class_1 = X[t == 1]
class_2 = X[t == 2]

mu_1 = np.mean(class_1, axis=0)
mu_2 = np.mean(class_2, axis=0)
Sigma_1 = np.cov(class_1.T)
Sigma_2 = np.cov(class_2.T)
w = np.linalg.solve(Sigma_1, mu_1 - mu_2)
w_0 = -0.5 * (mu_1 @ np.linalg.solve(Sigma_1, mu_1) - mu_2 @ np.linalg.solve(Sigma_2, mu_2)) - np.log(0.5 / 0.5)

x = np.linspace(-5, 5, 100)
y = -(w[0] * x + w_0) / w[1]
plt.plot(x, y, 'r-', label='LDA Decision Boundary')
plt.legend()
plt.show()
````
